```python
def create_login_screen():
    login_sc = Tk()
    login_sc.title('Login')
    login_sc.resizable(False, False)
    login_sc.geometry('800x400+500+400')
    
    # Criando Componentes
    lbl_username = Label(login_sc, text = 'Usuário').pack()
    tb_username = Entry(login_sc, text = 'Usuário', width = 30).pack()
    lbl_password = Label(login_sc,text = 'Senha').pack()
    tb_password = Entry(login_sc, show = '*', width = 20).pack()
    bt_login = Button(login_sc,text = 'Entrar').pack()
    bt_new_user = Button(login_sc,text = 'Cadastrar Novo').pack()
    
    #Incluindo Componetes no form
    
    return login_sc
    #login_sc.mainloop()
```

```python
login_sc = create_login_screen()
login_sc.mainloop()
```
